# ADNI DATA

```python
df = pd.read_csv('../datasets/adni_data/ADNI_sMRI.csv')
print(df.shape)
df
```

```python
"""
Check for NAN values
"""
df.isnull().sum().sort_values(ascending=False)
```

```python
# drop all columns that have all NAN values (left with 36 features)
df = df.dropna(axis=1, how='all')
df
```

```python
df['TotalICVolume'].describe()

# This can be dropped potentially
```

```python
# feature unnamed: 25???
df.drop(['Unnamed: 25'], axis=1)
```

```python
# how many columns have some nan values
null_rows = df.isnull().any(axis=1)
total_null = sum(null_rows)
print("Rows containing at least one NAN values: " + str(total_null))
```

```python
# Extract a list of non nan clolumns
mri_features = list(df.columns[1:])
file = open('../preprocessed_data/ADNI_MRI_BRAIN_REGIONS.csv', 'w')
for f in mri_features:
    file.write(f+"\n")
file.close()
```

```python
# Write to new preprocessed file
df.to_csv('../preprocessed_data/ADNI_MRI.csv', encoding='utf-8', index=False)
df
```

```python
# Normalisation of values
from sklearn.preprocessing import MinMaxScaler

# TODO: feature_range=(min, max) find appropriate range for brain volume
scaler = MinMaxScaler()
df_id = pd.DataFrame(df['ID'], columns=['ID'])
df_scaled = scaler.fit_transform(df.drop('ID', axis=1))
df_scaled = pd.DataFrame(df_scaled, columns=df.columns[1:])
df_scaled = df_id.join(df_scaled, how='outer')
# print(f"max: {scaler.data_max_}")
df_scaled
```

```python
df_scaled.to_csv('../preprocessed_data/ADNI_MRI_SCALED.csv', encoding='utf-8', index=False)
```
