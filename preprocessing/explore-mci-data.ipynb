```python
%matplotlib inline
import matplotlib.pyplot as plt

#Load libraries for data processing
import pandas as pd
import numpy as np
# from scipy.stats import norm
import seaborn as sns # visualization


plt.rcParams['axes.titlesize'] = 'large'
```

#### Descriptive Statistics

```python
data = pd.read_csv('mci-clean-data.csv', index_col=False,)
data.head()
```

```python
data = data.drop(['AD'], axis=1)
data.describe()
```

The skew result show a positive (right) or negative (left) skew. Values closer to zero show less skew. 

```python
data.skew()
```

### Combined Features

2093 observations indicating a patient has MCI and 1347 show normal cognition

```python
label_split = data.groupby('MCI', axis=0)
pd.DataFrame(label_split.size(), columns=['# of observations'])
```

```python
y = data.MCI
list = ['MCI']
data = data.drop(list,axis = 1 )
data.head()
```

Understanding features through various scoring techniques

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# find best scored 5 features
select_feature = SelectKBest(chi2, k=5).fit(x_train, y_train)
```

```python
print('Score list:', select_feature.scores_)
print('Feature list:', x_train.columns)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Create the RFE object and rank each pixel
rf_2 = RandomForestClassifier()      
rfe = RFE(estimator=rf_2, n_features_to_select=10, step=1)
rfe = rfe.fit(x_train, y_train)
```

```python
print('Chosen best 10 feature by rfe:',x_train.columns[rfe.support_])
```

### Neuro Features

```python
data = pd.read_csv('data/Neuro_MCI_NC.csv', index_col=False)
```

```python
y = data.MCI                     
list = ['MCI', 'AD']
data = data.drop(list,axis = 1 )
data.head()
```

```python
# # plot correlation matrix
plt.style.use('fivethirtyeight')
sns.set_style("white")

# Compute the correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
_, ax = plt.subplots(figsize=(8, 8))
plt.title('AD Feature Correlation')

# Generate a custom diverging colormap
cmap = sns.diverging_palette(260, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, vmax=1.2, square='square', cmap=cmap, mask=mask, ax=ax,annot=True, fmt='.2g',linewidths=2)
```

```python
sns.jointplot(data.loc[:,'MMSE'], data.loc[:,'ADAS11'], kind="reg")
```

```python
neuro_hist = data.hist(bins=10, figsize=(15, 10), grid=False)
```
